#### Programming approach

<pre>
We propose a 5 stage approach to solve any problem efficiently, 

Stage 1: [Requirment Analysis]
i)   Describe the requirement clearly including the stake holders and their actions. 
ii)  Conditions specified and that needs to be followed for solving given problem. (could include deployment environment details)
iii) Input provided
iv)  Output expected
v)   Test Cases 
{Best practices, always get your stage 1 reviewed and approved by customer before going to the next stage.}


Stage 2: [Design]
Break down problem/requirement into parts. 
(basing on stake holders, departments, change in state, location etc) 
[Part 1, Part 2 , Part 3...,Part n]

Stage 3: [Design]
Take each part from the above breakdown and write an algorithm for each part
Part 1: 
<algorithm to solve part 1>
Part 2: 
<algorithm to solve part 2>
.
.
.
Part n:
<algorithm to solve part n>

Linking/Integration code snippets for above parts { if required as per requirement }

Stage 4: [Coding]
Take each part algorithm and convert into the language in which you would deploy the solution 
{Java, JS, Python, .Net, any language}

Stage 5: [Deployment]
Test the above solution to see if its providing required solution, you can also prepare set of testcases (from Stage 1) or more and test it, if its not solving/satisfying given problem/requirement. Go to Stage 2 and break it down further or see if you missed any part of the requirement during breakdown. 
Deploy the final solution of the program in the target environment as per requirement and get it test by the customers. 


</pre>
